```python
import pandas as pd

df = pd.read_csv(r"C:\Users\user\Downloads\Food_Delivery_Times.csv")
df.head()

```

```python
print(df.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data
weather_data = df.groupby("Weather")["Delivery_Time_min"].mean().reset_index()
weather_data = weather_data.sort_values(by="Delivery_Time_min", ascending=False)

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x="Weather", y="Delivery_Time_min", hue="Weather", data=weather_data, palette="coolwarm", legend=False)
plt.xlabel("Weather Condition")
plt.ylabel("Average Delivery Time (min)")
plt.title("Impact of Weather on Delivery Time")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python

traffic_data = df.groupby("Traffic_Level")["Delivery_Time_min"].mean().reset_index()
traffic_data = traffic_data.sort_values(by="Delivery_Time_min", ascending=False)

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x="Traffic_Level", y="Delivery_Time_min", hue="Traffic_Level", data=traffic_data, palette="coolwarm", legend=False)
plt.xlabel("Traffic Condition")
plt.ylabel("Average Delivery Time (min)")
plt.title("Impact of Traffic on Delivery Time")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
time_data = df.groupby("Time_of_Day")["Delivery_Time_min"].mean().reset_index()
time_data = time_data.sort_values(by="Delivery_Time_min", ascending=False)

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x="Time_of_Day", y="Delivery_Time_min", hue="Time_of_Day", data=time_data, palette="coolwarm", legend=False)
plt.xlabel("Time of day")
plt.ylabel("Average Delivery Time (min)")
plt.title("Impact of time on Delivery Time")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
courier_data = df.groupby("Courier_Experience_yrs")["Delivery_Time_min"].mean().reset_index()
courier_data = courier_data.sort_values(by="Delivery_Time_min", ascending=False)

# Visualization
plt.figure(figsize=(8,5))
sns.barplot(x="Courier_Experience_yrs", y="Delivery_Time_min", hue="Courier_Experience_yrs", data=courier_data, palette="coolwarm", legend=False)
plt.xlabel("Courier_Experience_yrs")
plt.ylabel("Average courier experience (min)")
plt.title("Impact of experience on Delivery Time")
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.show()
```

```python
df["Delivery_Status"] = df["Delivery_Time_min"].apply(lambda x: "Late" if x > df["Delivery_Time_min"].mean() else "On-Time")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of 'Late' and 'On-Time' deliveries
status_counts = df["Delivery_Status"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=["#ff9999", "#66b3ff"], startangle=140)
plt.title("Delivery Status Distribution")
plt.show()
```
